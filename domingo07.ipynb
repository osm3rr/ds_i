```python
import pandas as pd
```

```python
df_titanic = pd.read_csv('titanic.csv')
```

```python
df_titanic.head(3)
```

```python
df_age_category = df_titanic[["Name", "Age", "Sex", "Fare"]]
```

```python
df_age_category.head()
```

```python
df_age_category["age_category"] = df_age_category["Age"]
```

```python
df_age_category.head()
```

```python
# funcion para definir cateorias de edad
# niño < 12, adolescente 12-17, adulto 18-55, adulto mayor 56+
def category_f(age):
    if (age > 0) and (age <= 1):
        return 'baby' 
    elif (age > 1) and (age < 12):
        return 'child'
    elif (age >= 12) and (age < 18):
        return 'teens'
    elif (age >= 18) and (age <= 55 ):
        return 'adult'
    else:
        return 'senior'
```

```python
df_age_category.info()
```

```python
# identifico la cantidad de filas de edad sin valor
df_age_category["Age"].isna().sum()
```

```python
# sacar los valores nulos de la columna edad
df_age_category = df_age_category.dropna()
```

```python
df_age_category.info()
```

```python
df_age_category["age_category"] = df_age_category["age_category"].apply(category_f)
```
